The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

# Start coding here!
df = pd.read_csv(
    "C:/Users/bdfba/AppData/Local/Temp/6044f537-8f4c-450e-8c07-4419bcbd631b_workspace(2).zip.workspace(2).zip/workspace/data/nobel.csv"
)

print(df.columns)

```

```python
# Top gender using groupby
top_gender = df.groupby('sex').size().sort_values(ascending=False).index[0]

# Top country using groupby
top_country = df.groupby('birth_country').size().sort_values(ascending=False).index[0]
```

```python

df['usa_born_winner'] = df['birth_country'] == 'United States of America'
df['decade'] = (np.floor(df['year'] / 10) * 10).astype(int)
prop_usa_winners = df.groupby('decade', as_index=False)['usa_born_winner'].mean()

# Identify the decade with the highest proportion of US-born winners
max_decade_usa = prop_usa_winners[prop_usa_winners['usa_born_winner'] == prop_usa_winners['usa_born_winner'].max()]['decade'].values[0]
print(max_decade_usa)
```

```python

# Filter female winners
df_female = df[df['sex'] == 'Female']

# Total winners per decade and category
decade_cat_counts = df.groupby(['decade', 'category']).size()

# Female winners per decade and category
female_counts = df_female.groupby(['decade', 'category']).size()

# Ratio
female_ratio = female_counts / decade_cat_counts

# Get decade and category with highest ratio
max_decade, max_category = female_ratio.sort_values(ascending=False).index[0]

# Store as dictionary
max_female_dict = {int(max_decade): max_category}

print(max_female_dict)
```

```python
# Filter female laureates
df_female = df[df['sex'] == 'Female']

# Sort by year to get the earliest winner
first_woman = df_female.sort_values('year').iloc[0]

# Store answers as strings
first_woman_name = first_woman['full_name']
first_woman_category = first_woman['category']

print(first_woman_name)
print(first_woman_category)
```

```python
# Count how many times each full name appears
name_counts = df.groupby('full_name').size()

# Select names with more than one Nobel Prize
repeat_list = name_counts[name_counts > 1].index.tolist()

print(repeat_list)
```
